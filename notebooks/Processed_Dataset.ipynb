```python
plt.figure(figsize=(15, 8))
train_df.hist(bins=30, figsize=(15, 8), color='blue', edgecolor='black')
plt.suptitle('Histograms of Numeric Features - Train Dataset')
plt.show()
```

```python
categorical_cols = train_df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=col, data=train_df, palette='Set2')
    plt.title(f'Distribution of Categorical Feature: {col}')
    plt.show()
```

```python
numeric_df = train_df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Correlation Matrix - Train Dataset')
plt.show()
```

```python
numeric_cols = train_df.select_dtypes(include=[np.number])
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))).sum()
print("\nOutliers detected in Train Dataset:")
print(outliers)
```

```python
train_df[numeric_cols.columns] = train_df[numeric_cols.columns].fillna(train_df[numeric_cols.columns].median())
```

```python
train_df_encoded = pd.get_dummies(train_df, drop_first=True)
```

```python
train_df_encoded.to_csv('processed_train_df.csv', index=False)
```
